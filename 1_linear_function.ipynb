```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-25, 25, step=0.1)
```

```python
def get_base_fig(x_min = -10, x_max = 10, y_min = -8, y_max = 8):
    fig, ax = plt.subplots(figsize = (11, 8))
    ax.hlines(y=0, color='gray', xmin=-100, xmax = 100, linestyles='dashed', alpha=0.7)
    ax.vlines(x=0, color='gray', ymin=-100, ymax = 100, linestyles='dashed', alpha=0.7)
    ax.set_xlim(x_min, x_max)
    ax.set_ylim(y_min, y_max)
    ax.set_xlabel("x")
    ax.set_ylabel("f(x)")
    ax.set_aspect('equal')
    ax.grid(alpha=0.3, color = 'gray')                                    
    return fig, ax
```

# Straight line equation

A straight line is a function given by the equation:
$$f(x) = a \cdot x + b$$
or a graph 
$$y = a \cdot x + b$$

where $a$ and $b$ are constant parameters. $a$ is called the gradient of the straight line, $b$ is called the intercept of the straight line.

```python
fig, ax = get_base_fig()

ax.plot(x, x + 3, color = 'blue')
ax.set_title("$f(x) = x+3$")
plt.show()
```

```python
fig, ax = get_base_fig()
ax.plot(x, 2*x + 1, color = 'blue')
ax.set_title("$f(x) = 2x+1$")
plt.show()
```

```python
fig, ax = get_base_fig()
ax.plot(x, 0.5*x - 1.5, color = 'blue')
ax.set_title("$f(x) = 0.5x - 1.5$")
plt.show()
```

```python
fig, ax = get_base_fig()
ax.plot(x, x - 4, color = 'blue')
ax.set_title("$f(x) = -x - 4$")
plt.show()
```

# Straight line formed from two points:

A straight line can be created by joining two points with coordinates: $(x_1, y_1)$ and $(x_2, y_2)$. The gradient and the intercept of the resulting line are given by:

$$ a = \frac{y_2 - y_1}{x_2 -x_1}$$
$$b = \frac{y_1x_2-y_2x_1}{x_2 -x_1}$$


### Proof: 


Inserting $(x_1, y_1)$ and $(x_2, y_2)$ to the straingt line formula $f(x) = a \cdot x + b$ gives the following set of equations:

$$\begin{equation*}
  \left\{
    \begin{aligned}
      & f(x_1) = y_1 = a \cdot x_1 + b \\
      & f(x_2) = y_2 = a \cdot x_2 + b
    \end{aligned}
  \right.
\end{equation*}$$


We start with subtracting the first equation from the second equation to obtain the gradient $a$:

$$ y_2 - y_1 = (a\cdot x_2 + b) - (a\cdot x_1 + b) $$
$$ y_2 - y_1 = a \cdot (x_2 -x_1)$$

$$ a = \frac{y_2 - y_1}{x_2 -x_1}$$

Inserting $ a = \frac{y_2 - y_1}{x_2 -x_1}$ to the first equation gives the intercept $b$:


$$b = y_1 - a\cdot x_1 = y_1 - \frac{y_2 - y_1}{x_2 -x_1} = \frac{y_1(x_2-x_1)-(y_2-y_1)x_1}{x_2 -x_1}
= \frac{y_1x_2-y_1x_1-y_2 x_1 + y_1x_1}{x_2 -x_1} = \frac{y_1x_2-y_2x_1}{x_2 -x_1}
$$

```python
def create_line_from_points(x1, y1, x2, y2):
    
    fig, ax = get_base_fig()
    ax.scatter([x1, x2], [y1, y2], color='red', s=20)
    ax.text(x1, y1, f"A=({x1}, {y1})", size=12, ha='right', va='bottom')
    ax.text(x2, y2, f"B=({x2}, {y2})", size=12, ha='right', va='bottom')
    
    if x1 == x2:
        ax.vlines(x1, ymin=-100, ymax=100, color = 'black')
    else:
        a = (y2-y1)/(x2-x1)
        b = (y1*x2 - x1*y2)/(x2-x1)
        ax.plot(x, a*x+b, color = 'black')
        ax.set_title(f"f(x) = {round(a, 2)} x + {round(b, 2)}")

    plt.show()
```

```python
create_line_from_points(x1=4, y1=2, x2=6, y2=3)
```

# Intersection points of the straight line with the x-axis and y-axis

Taking the straight line function $f(x) = ax+b$ at $x=0$ gives the intersection point with the y-axis
$$f(0) = a\cdot 0+b \rightarrow y=b$$

$$\begin{equation*}
  \left\{
    \begin{aligned}
      & x = 0 \\
      & y = b
    \end{aligned}
  \right.
\end{equation*}$$


Taking the straight line function $f(x) = ax+b$ at $y=0$ gives the intersection point with the x-axis

$$f(x) = 0 = a\cdot x+b \rightarrow x = -\frac{b}{a}$$


$$\begin{equation*}
  \left\{
    \begin{aligned}
      & x = -\frac{b}{a} \\
      & y = 0
    \end{aligned}
  \right.
\end{equation*}$$

```python
def get_intersection_points(a, b):
    
    x1 = 0
    y1 = round(b, 2)
    x2 = round(-b/a, 2)
    y2 = 0
    
    fig, ax = get_base_fig()
    ax.plot(x, a*x+b, color = 'black')
    ax.scatter([x1, x2], [y1, y2], color='red')
    ax.text(x1, y1, f"A=({x1}, {y1})", size=12, ha='right', va='bottom')
    ax.text(x2, y2, f"B=({x2}, {y2})", size=12, ha='right', va='bottom')
    ax.set_title(f"f(x) = {a} x + {b}")
    plt.show()
```

```python
get_intersection_points(a=2, b=5)
```

```python
get_intersection_points(a=0.7, b=-2)
```

# Parallel lines

Parallel lines are coplanar straight lines that do not intersect at any point. Parallel lines have the same gradient.

```python
b_list = [-20, -10, -5, -2, 2, 5, 10]

fig, ax = get_base_fig()
ax.plot(x, 2*x, color = 'purple', label = f"f(x) = 2x")
for i, b in enumerate(b_list):
    ax.plot(x, 2*x + b, color = 'blue', alpha = 1-i/10, label = f"f(x) = 2x + {b}")
ax.legend(loc='upper left')
ax.set_title("Lines parallel to the line $f(x) = 2\cdot x$")
plt.show()
```

```python
b_list = [-20, -10, -5, -2, 2, 5, 10]

fig, ax = get_base_fig()
ax.plot(x, 0.5*x, color = 'purple', label = "f(x) = 0.5 x")
for i, b in enumerate(b_list):
    ax.plot(x, 0.5*x + b, color = 'blue', alpha = 1-i/10, label = f"f(x) = 0.5x + {b}")
ax.legend(loc='upper left')
ax.set_title("Lines parallel to the line $f(x) = 0.5 \cdot x$")
plt.show()
```

# Perpendicular lines


A line is said to be perpendicular to another line if the two lines intersect at a right angle. In the two-dimensional plane, right angles can be formed by two intersected lines if the product of their gradients equals −1. 

So two lines given by linear functions 
$$ y = f(x) = a_1 x + b_1$$
$$ y = f(x) = a_2 x + b_2$$

are perpendicular if:
$$a_1 \cdot a_2 = -1$$ 

```python
b_list = [-10, -5, 0, -2, 2, 5]

fig, ax = get_base_fig()
ax.plot(x, 5*x, color = 'purple', label = "f(x) = 5x")
for i, b in enumerate(b_list):
    ax.plot(x, -0.2*x + b, color = 'blue', alpha = 1-i/8, label = f"f(x) = -0.2 x + {b}")
ax.legend(loc='lower left')
ax.set_title("Lines perpendicular to the line $f(x) = 5 \cdot x$")
plt.show()
```
