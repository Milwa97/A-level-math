```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-25, 25, step=0.02)

x_neg = np.arange(-25, 0, step=0.02)
x_pos = np.arange(0.02, 25, step=0.02)
```

```python
def get_base_fig(x_min = -8, x_max = 8, y_min = -10, y_max = 25):
    fig, ax = plt.subplots(figsize = (11, 6))
    ax.hlines(y=0, color='gray', xmin=-100, xmax = 100, linestyles='dashed', alpha=0.7)
    ax.vlines(x=0, color='gray', ymin=-100, ymax = 100, linestyles='dashed', alpha=0.7)
    ax.set_xlim(x_min, x_max)
    ax.set_ylim(y_min, y_max)
    ax.set_xlabel("x")
    ax.set_ylabel("f(x)")
    ax.grid(alpha=0.3, color = 'gray')                                       
    return fig, ax
```

### Exercise 2
a) On the same axes sketch the curves given by $y = x^2 (x − 3)$ and $y = \frac{2}{x}$. 

b) Explain how your sketch shows that there are only two real solutions to the equation
$$ x^3(x − 3) = 2 $$

```python
fig, ax = get_base_fig()
ax.plot(x, (x+1)**3, color = 'green', label = "$f(x) = (x+1)^3$")
ax.plot(x, 3*x*(x-1), color = 'blue', label = "$f(x) = 3x(x-1)$")
ax.legend(loc = 'upper right') 
plt.show()
```

### Exercise 4


a) On the same axes sketch the curves given by $y = \frac{1}{x}$ and $y = −x(x − 1)^2$.

b) Explain how your sketch shows that there are no real solutions to the equation
$$ 1 + x^2(x − 1)^2 = 0$$

```python
fig, ax = get_base_fig()
ax.plot(x, -x*(x-1)**2, color = 'green', label = r"$f(x) = -x(x-1^2)$")
ax.plot(x_neg, 1/x_neg, color = 'blue', label = r"$f(x) = \frac{1}{x}$")
ax.plot(x_pos, 1/x_pos, color = 'blue')

ax.legend(loc = 'upper right') 
plt.show()
```

### Exercise 5

a) On the same axes sketch the curves given by $y = x^2 (x + a)$ and $y = \frac{b}{x}$ where a and b are both positive constants. 

b) Using your sketch, state, giving a reason, the number of real solutions to the equation 
$$ x^4 + ax^3 – b = 0. $$

```python
a_tab = [0.5, 1, 2, 5, 10]
b_tab = [0.5, 1, 2, 5, 10]
alpha_tab = [1, 0.8, 0.6, 0.4]

fig, ax = get_base_fig()

for alpha, a in zip(alpha_tab,a_tab):
    ax.plot(x, x**2 * (x+a), color = 'green', alpha=alpha, label = r"$f(x) = x^2 (x+a)$")

for alpha, b in zip(alpha_tab, b_tab):
    ax.plot(x_neg, b/x_neg, color = 'blue', alpha=alpha, label = r"$f(x) = \frac{b}{x}$")
    ax.plot(x_pos, b/x_pos, color = 'blue', alpha=alpha)

ax.legend(loc = 'upper right') 
plt.show()
```
