```python
def get_base_fig(x_min = -360, x_max = 360, y_min = -1.2, y_max = 1.2, step=30):

    degrees_ticks_positions = np.arange(x_min, x_max, step=step)
    degrees_ticks = [str(degree) + '$^o$' for degree in degrees_ticks_positions]

    fig, ax = plt.subplots(figsize = (10, 6))
    ax.hlines(y=0, color='gray', xmin=-1000, xmax = 1000, linestyles='dashed', alpha=0.7)
    ax.vlines(x=0, color='gray', ymin=-15, ymax = 15, linestyles='dashed', alpha=0.7)
    ax.set_xlim(x_min, x_max)
    ax.set_ylim(y_min, y_max)
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_xticks(degrees_ticks_positions, degrees_ticks)
    ax.tick_params(axis='x', rotation=45)
    ax.grid(alpha=0.3, color = 'gray')                                       
    return fig, ax
```

```python
def show_rendered_plots(fig, ax, x_min, x_max, step, show_fig = True):
    
    degrees_ticks_positions = np.arange(x_min, x_max, step=step)
    degrees_ticks = [str(degree) + '$^o$' for degree in degrees_ticks_positions]

    for ax in axs:
        ax.set_xlim(x_min, x_max)
        ax.set_ylim(-1.2, 1.2)
        ax.set_xticks(degrees_ticks_positions, degrees_ticks)
        ax.tick_params(axis='x', rotation=45)
        ax.grid(alpha=0.3, color = 'gray')                                       
        ax.hlines(y=0, color='gray', xmin=-1000, xmax = 1000, linestyles='dashed', alpha=0.7)
        ax.vlines(x=0, color='gray', ymin=-15, ymax = 15, linestyles='dashed', alpha=0.7)
        ax.set_xlabel("x")
        ax.set_ylabel("y")
        ax.legend(loc='upper right')
    
    if show_fig:
        plt.show()
```

# Radians and degrees:

The radian, denoted by the symbol rad, is the unit of angle in the International System of Units (SI) and is the standard unit of angular measure used in many areas of mathematics. 

One radian is defined as the angle subtended from the center of a circle which intercepts an arc equal in length to the radius of the circle. More generally, the magnitude in radians of a subtended angle is equal to the ratio of the arc length to the radius of the circle; that is, $ \theta ={\frac {s}{r}} $, where $\theta$ is the subtended angle in radians, $s$ is arc length, and $r$ is radius. The rotation angle ($360^o$) corresponding to one complete revolution is the length of the circumference divided by the radius, which is $\frac {2\pi r}{r}$, or $2\pi$. Thus, $2\pi$ radians is equal to $360^o$. 
Source: [wikipedia](https://en.wikipedia.org/wiki/Radian)

Conversion from radians to degrees:

|degrees | radians|
|------------|----------------------------------------|
| 0    |  0 |
| 30   | $\frac{1}{6}\pi$   |
| 45   | $\frac{1}{4}\pi$   |
| 60   | $\frac{1}{3}\pi$   |
| 90   | $\frac{1}{2}\pi$   |
| 180   | $\pi$   |
| 270   | $\frac{3}{2}\pi$   |
| 360   | $2\pi$   |

# Trigonometric functions

The trigonometric functions are real functions which relate an angle of a right-angled triangle to ratios of two side lengths. 


If the acute angle θ is given, then any right triangles that have an angle of θ are similar to each other. This means that the ratio of any two side lengths depends only on θ. Thus these six ratios define six functions of θ, which are the trigonometric functions. In the following definitions, the hypotenuse is the length of the side opposite the right angle, opposite represents the side opposite the given angle θ, and adjacent represents the side between the angle θ and the right angle.

$$ 
\begin{aligned}
        &\sin(\alpha )={\frac {\textrm {opposite}}{\textrm {hypotenuse}}}\\
        &\cos(\alpha )={\frac {\textrm {adjacent}}{\textrm {hypotenuse}}}\\
        &\tan(\alpha )={\frac {\textrm {opposite}}{\textrm {adjacent}}}\\
        &\cot(\alpha )={\frac {\textrm {adjacent}}{\textrm {opposite}}}\\
\end{aligned}$$

# Sine vs cosine

```python
x = np.linspace(-360, 360, 1001)

fig, ax = get_base_fig()
ax.plot(x, sin(x), color = 'blue', label='$f(x)=\sin(x)$')
ax.plot(x, cos(x), color = 'green', label='$f(x)=\cos(x)$')
ax.legend(loc='upper right')
ax.set_title("sin(x) vs cos(x)")
plt.show()
```

# Trigonometric one:

For any real $x$ the following property holds:

$$\sin(x)^2 + \cos(x)^2 = 1$$

```python
x = np.linspace(-270, 270, 1001)

fig, axs = plt.subplots(figsize = (11, 9), nrows=3, ncols=1, sharex=True)
fig.suptitle("Trigonometric one: $\sin(x)^2 + \cos(x)^2 = 1$")

axs[0].plot(x, sin(x), color = 'blue', label = r"$f(x) = \sin(x)$")
axs[1].plot(x, cos(x), color = 'green', label = r"f$(x) = \cos(x)$")
axs[2].plot(x, sin(x)**2, color = 'blue', label='$f(x)=\sin(x)^2$')
axs[2].plot(x, cos(x)**2, color = 'green', label='$f(x)=\cos(x)^2$')
axs[2].plot(x, cos(x)**2 + sin(x)**2, color = 'red', label='$f(x)=\sin(x)^2 + \cos(x)^2$')

show_rendered_plots(fig, axs, x_min=-270, x_max=270, step=30, show_fig=False)
axs[2].legend(loc='lower right')
plt.show()
```
