```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-25, 25, step=0.02)
```

```python
def get_base_fig(x_min = -10, x_max = 10, y_min = -2, y_max = 20):
    fig, ax = plt.subplots(figsize = (11, 6))
    ax.hlines(y=0, color='gray', xmin=-100, xmax = 100, linestyles='dashed', alpha=0.7)
    ax.vlines(x=0, color='gray', ymin=-100, ymax = 100, linestyles='dashed', alpha=0.7)
    ax.set_xlim(x_min, x_max)
    ax.set_ylim(y_min, y_max)
    ax.set_xlabel("x")
    ax.set_ylabel("f(x)")
    ax.grid(alpha=0.3, color = 'gray')                                       
    return fig, ax
```

# Quadratic function:

A quadratic polynomial is a polynomial of degree two in one or more variables. In this notebook only univariate (single-variable)  polynomials are considered. A quadratic function is the polynomial function defined by a quadratic polynomial:
$$ f(x) = a\cdot x^2 + b \cdot x + c$$


A univariate quadratic function is usually presented in one of the following forms:

## standard form:
$$ f(x) = a\cdot x^2 + b \cdot x + c$$
$a,b,c$ are real constants


##  vertex form
$$ f(x) = a \cdot (x+p) ^2 + q$$
$a, p,q $ are real constants


## factored form
$$ f(x) = a\cdot (x-r_1) \cdot (x-r_2)$$
$a,r_1, r_2$ are real constants, $r_1, r_2$ are the roots of the quadratic function and the solutions of the corresponding quadratic equation.



# Standard form

$$ f(x) = a\cdot x^2 + b \cdot x + c$$

$a,b,c$ are real constants, 

* $a$ coefficient stretches and flips the function
* $b$ coefficient moves the function along both the horizontal and vertical axis 
* $c$ coefficient moves the function along the vertical axis

# Conversion from standard formula to factored form*


### * Conversion from standard formula to factored form is not always possible!

### standard formula: $ f(x) = a\cdot x^2 + b \cdot x + c$
###  factored form: $ f(x) = a\cdot (x-r_1) \cdot (x-r_2) $

$ f(x) = a\cdot x^2 + b \cdot x + c = a\left( x^2 + \frac{b}{a} x + \frac{c}{a} \right) 
  = a\left(x- \frac{-b + \sqrt{b^2-4ac}}{2a} \right) \cdot \left(x + \frac{-b + \sqrt{b^2-4ac}}{2a} \right) 
$

so:

$$r_1 = \frac{-b + \sqrt{b^2-4ac}}{2a} = \frac{-b + \sqrt{\Delta}}{2a} $$
and
$$r_2 = \frac{-b - \sqrt{b^2-4ac}}{2a} = \frac{-b - \sqrt{\Delta}}{2a}$$

where
$\Delta = b^2 - 4ac$ is called the discriminant.

The value of the discriminant shows how many roots f(x) has:
* if $\Delta = b^2  4ac > 0$ then f(x) has two distinct real roots
* if $\Delta = b^2  4ac = 0$ then f(x) has one repeated root
* if $\Delta = b^2  4ac < 0$ then f(x) has has no real roots





```python
def standard_formula_to_factorial_form(a,b,c, as_string=True):
    discriminant = b**2 - 4*a*c
    if discriminant > 0:
        r1 = np.round((-b - np.sqrt(discriminant))/(2*a), 3)
        r2 = np.round((-b + np.sqrt(discriminant))/(2*a), 3)
        if as_string:
            return f"a = {a}, r1 = {r1}, r2 = {r2}"
        else:
            return a, r1, r2
    elif discriminant == 0:
        r0 = np.round(-b/(2*a), 3)
        if as_string:
            return  f"a = {a}, r_0 = {r0}"
        else:
            return a, r0, r0
    else:
        if as_string:
            return "f(x) has has no real roots and no factorial form"
        else:
            return None
    
def factorial_to_standard_form(a, r1, r2, as_string=True):
    b = np.round(-a*(r1+r2), 3)
    c = np.round(a*r1*r2, 3)
    if as_string:
        return f"a = {a}, b = {b}, c = {c}"
    else:
        return a, b, c
```

```python
factorial_to_standard_form(a=1, r1=3, r2=2)
```

```python
factorial_to_standard_form(a=3, r1=2, r2=1.5)
```

```python
standard_formula_to_factorial_form(a=1, b=-5, c=6)
```

# Conversion from standard formula to vertex form

### standard formula: $ f(x) = a\cdot x^2 + b \cdot x + c$
###  vertex form: $ f(x) = a \cdot (x+p) ^2 + q$

$ f(x) = a\cdot x^2 + b \cdot x + c = a \left( x^2 + \frac{b}{a}x \right) +c
= a \cdot \left( x+ \frac{b}{2a}\cdot 2x + \frac{b^2}{4a^2}- \frac{b^2}{4a^2} \right) +c 
= a \cdot \left( x+ \frac{b}{2a} \right)^2 + \left(c-\frac{b^2}{4a}\right) = a \cdot (x+p) ^2 + q$

so:

$$p=\frac{b}{2a}$$
$$q=c-\frac{b^2}{4a}$$




##  Examples:

$$ f(x) = x^2+4x \quad \rightarrow \quad f(x) = (x+2)^2 - 4$$

$$ f(x) = x^2-6x \quad \rightarrow \quad f(x) = (x-3)^2 - 9$$

$$ f(x) = 2 x^2+16x \quad \rightarrow \quad f(x) = 2\cdot(x+4)^2 - 32$$

$$ f(x) = 3 x^2-24x \quad \rightarrow \quad f(x) = 3\cdot (x-4)^2 - 48$$

$$ f(x) = 2 x^2+8x +1 \quad \rightarrow \quad f(x) = 2\cdot(x+2)^2 - 7$$

$$ f(x) = -4x^2-16x + 10 \quad \rightarrow \quad f(x) = -4\cdot (x+2)^2 +26$$


```python
def standard_formula_to_vertex(a, b, c, as_string=True):
    p = np.round(b / (2*a), 3)
    q = np.round(c - (b**2/(4*a)), 3)
    if as_string:
        return f"a = {a}, p = {p}, q = {q}"
    else:
        return a, p, q

def vertex_to_standard_formula(a, p, q, as_string=True):
    b = 2*a*p
    c = a*p**2 + q
    if as_string:
        return f"a = {a}, b = {b}, c = {c}"
    else:
        return a, b, c
```

```python
standard_formula_to_vertex(a=-4, b=-16, c=10)
```

```python
vertex_to_standard_formula(a=-4, p=2, q=26)
```

# All in one

```python
def show_quadratic_function_info(a, b, c, y_min = -5, y_max = 15):
            
    roots = standard_formula_to_factorial_form(a, b, c, as_string=False)
    a, p, q = standard_formula_to_vertex(a, b, c, as_string=False)
    
    fig, ax = get_base_fig(y_min =y_min, y_max=y_max)
    ax.set_title(f"f(x) = {a} x^2 + {b}x +{c}")
    ax.plot(x, a*x**2+b*x+c, color = 'black')
    ax.scatter([-p], [q], color='purple', s=25, label=f"turning point at ({-p},{q})")
    if roots:
        a, r1, r2 = roots
        ax.scatter([r1, r2], [0,0], color='red', s=25, label=f'roots r1 = {r1}, r2 = {r2}')       
    ax.legend(loc='lower right')    
    plt.show()
```

```python
show_quadratic_function_info(a=-1, b=6, c=7)
```
