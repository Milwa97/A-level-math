```python
import matplotlib.pyplot as plt
from mycolorpy import colorlist as mcp
import numpy as np

t = np.linspace(-2*np.pi, 2*np.pi, 1001)
cmap = 'gist_rainbow'
color_list = mcp.gen_color(cmap=cmap, n=10)[::-1]
```

```python
def get_base_fig(x_min = -5, x_max = 5, y_min = -5, y_max = 5, step=1):
    fig, ax = plt.subplots(figsize = (11, 6))
    ax.hlines(y=0, color='gray', xmin=-10, xmax = 100, linestyles='dashed', alpha=0.7)
    ax.vlines(x=0, color='gray', ymin=-10, ymax = 100, linestyles='dashed', alpha=0.7)
    ax.set_xlim(x_min, x_max)
    ax.set_ylim(y_min, y_max)
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_xticks(np.arange(x_min, x_max, step = 1))
    ax.set_yticks(np.arange(y_min, y_max, step = step))
    ax.set_aspect('equal')
    ax.grid(alpha=0.3, color = 'gray')                                       
    return fig, ax
```

# Conic section

In mathematics, a conic section, quadratic curve or conic is a curve obtained as the intersection of the surface of a cone with a plane. The three types of conic section are the hyperbola, the parabola, and the ellipse; the circle is a special case of the ellipse, though historically it was sometimes called a fourth type. 

In analytic geometry, a conic may be defined as a plane algebraic curve of degree 2; that is, as the set of points whose coordinates satisfy a quadratic equation in two variables which can be written in the form 
$$ a\cdot x^2 + b\cdot x\cdot y + c\cdot y^2 + d\cdot x + e\cdot y + f = 0 $$

where $a$, $b$, $c$, $d$, $e$ and $f$ are real parameters

The geometric properties of the conic can be deduced from its equation. 


|conic section| equation| parametrization|
|------------|----------------------------------------|-----------------------------|
| circle      | $x^2 + y^2 =r^2 $                     |($r\cos \theta$, $r\sin\theta$)|
| ellipse     | $\frac{x^2}{a^2} + \frac{y^2}{b^2}=1$ |($a\cos \theta$, $b\sin\theta$)|
| parabola    | $y^2 = 4a\cdot x $                    |($at^2$, $2at$)|
| hyperbola   | $\frac{x^2}{a^2} - \frac{y^2}{b^2}=1 $|($\pm a\cosh\psi$, $b\sinh\psi$)|
|------------|----------------------------------------|------------|

# circle

$$ x^2 + y^2 = r^2 $$

where $r$ is called radius.

Circle parametrization:

$$
\left\{
    \begin{array}\\
        x  = r \cos{t}\\
        y  = r \sin{t}
    \end{array}
\right.
$$

```python
fig, ax = get_base_fig()
ax.plot(np.cos(t), np.sin(t), color = 'black', label = r"$x^2 + y^2 = 1$")
ax.legend(loc = 'lower right')
ax.set_title("Circle with radius $r=1$")
plt.show()
```

```python
r_list = [0.25, 0.5, 1, 2, 4]
fig, ax = get_base_fig()
for r, color in zip(r_list, color_list):
    ax.plot(r*np.cos(t), r*np.sin(t), color = color, label = r"$x^2 + y^2 =$" + str(r))
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
ax.set_title("Circle with radius $r$")
plt.show()
```

# Ellipse

Analytically, the equation of a standard ellipse centered at the origin with width $2a$ and height $2b$ is: 

$$ \frac{x^2}{a^2} + \frac{y^2}{b^2}=1 $$

where $a$ and $b$ are some constant parameters.

Ellipse parametrization:

$$
\left\{
    \begin{array}\\
        x = a \cdot \cos{t}\\
        y = b \cdot \sin{t}
    \end{array}
\right.
$$

parametric equation for eclipse with center at (u,v):

$$
\left\{
    \begin{array}\\
        x = u + a \cdot \cos{t}\\
        y = v + b \cdot \sin{t}
    \end{array}
\right.
$$


```python
a_list = [0.5, 1, 2, 4]
b = 1

fig, ax = get_base_fig()
for a, color in zip(a_list, color_list):
    ax.plot(a*np.cos(t), b*np.sin(t), color = color, label = f"a={a:.1f}, b={b:.1f}")
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
ax.set_title(r"Change $a$ parameter in ellipse equation: $\frac{x^2}{a^2} + \frac{y^2}{b^2}=1$")
plt.show()
```

```python
a = 1
b_list = [0.5, 1, 2, 4]

fig, ax = get_base_fig()
for b, color in zip(b_list, color_list):
    ax.plot(a*np.cos(t), b*np.sin(t), color = color, label = f"a={a:.1f}, b={b:.1f}")
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
ax.set_title(r"Change $b$ parameter in ellipse equation: $\frac{x^2}{a^2} + \frac{y^2}{b^2}=1$")
plt.show()
```

# Parabola

$$ y^2 = 4a\cdot x $$

where $a>0$ is some constant parameter.

Parabola parameterization:

$$
\left\{
    \begin{array}\\
        x = at^2\\
        y = 2at
    \end{array}
\right.
$$


```python
fig, ax = get_base_fig()
ax.plot(a*t**2, 2*a*t, color = 'black', label = r"$y^=4ax$")
ax.legend(loc='upper left')
ax.set_title("Parabola")
plt.show()
```

```python
a_list = [0.5, 1, 2, 4]
fig, ax = get_base_fig()
for a, color in zip(a_list, color_list):
    ax.plot(a*t**2, 2*a*t, color = color, label = f"a={a:.1f}")
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
ax.set_title(r"Change $a$ parameter in parabola equation: $y^2=4ax$")
plt.show()
```

# Hyperbola

$$ \frac{x^2}{a^2} - \frac{y^2}{b^2}=1 $$

Hyperbola parameterization:

$$
\left\{
    \begin{array}\\
        x = a \cdot \cosh{t}\\
        y = b \cdot \sinh{t}
    \end{array}
\right.
$$


```python
fig, ax = get_base_fig()
ax.plot(np.cosh(t), np.sinh(t), color = 'black', label = r"$x^2 - y^2 =1$")
ax.plot(-np.cosh(t), -np.sinh(t), color = 'black')
ax.legend(loc='lower right')
ax.set_title("Hyperbola")
plt.show()
```

```python
a_list = [0.5, 1, 2, 4]
b = 1

fig, ax = get_base_fig()
for a, color in zip(a_list, color_list):
    ax.plot(a*np.cosh(t), b*np.sinh(t), color = color, label = f"a={a:.1f}, b={b:.1f}")
    ax.plot(-a*np.cosh(t), -b*np.sinh(t), color = color)
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
ax.set_title(r"Change $a$ parameter in hyperbola equation: $\frac{x^2}{a^2} - \frac{y^2}{b^2}=1$")
plt.show()
```

```python
a = 1
b_list = [0.5, 1, 2, 5]

fig, ax = get_base_fig()

for b, color in zip(b_list, color_list):
    ax.plot(a*np.cosh(t), b*np.sinh(t), color = color, label = f"a={a:.1f}, b={b:.1f}")
    ax.plot(-a*np.cosh(t), -b*np.sinh(t), color = color)
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
ax.set_title(r"Change $b$ parameter in hyperbola equation: $\frac{x^2}{a^2} - \frac{y^2}{b^2}=1$")
plt.show()
```
