```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import Symbol, solve

x = np.arange(-25, 25, step=0.02)
```

```python
def get_base_fig(x_min = -8, x_max = 8, y_min = -10, y_max = 25):
    fig, ax = plt.subplots(figsize = (11, 6))
    ax.hlines(y=0, color='gray', xmin=-100, xmax = 100, linestyles='dashed', alpha=0.7)
    ax.vlines(x=0, color='gray', ymin=-100, ymax = 100, linestyles='dashed', alpha=0.7)
    ax.set_xlim(x_min, x_max)
    ax.set_ylim(y_min, y_max)
    ax.set_xlabel("x")
    ax.set_ylabel("f(x)")
    ax.grid(alpha=0.3, color = 'gray')                                       
    return fig, ax
```

# Cubic graphs

A cubic function is the polynomial function defined by a cubic polynomial, i.e., a polynomial of degree three in one or more variables. In this notebook only univariate (single-variable) polynomials are considered. 

A cubic function has the form $$f(x) = ax^3 + bx^2 + cx + d$$ where $a$, $b$, $c$ and $d$ are real numbers and $a\neq 0$. If for any value of $x_0$ $f(x_0)=0$, then $x_0$ is called a root of the function $f(x)$. In the other words, if $x_0$ is a root of the function $f(x)$, then the graph of $y = f(x)$ touches or crosses the x-axis at
the point $(x_0, 0)$. A cubic function can have at most three different roots. Sometimes, the same root occurs multiple times. In that case, that root is called a multiple root or repeated rood.


A univariate cubic function is usually presented in one of the following forms:


## standard form:
$$f(x) = ax^3 + bx^2 + cx + d$$
where $a$, $b$, $c$ and $d$ are real numbers and $a\neq 0$.


## factored form
$$ f(x) = a\cdot (x-r_1) \cdot (x-r_2)\cdot (x-r_3)$$
$a,r_1, r_2, r_3$ are the roots of the cubic function and the solutions of the corresponding cubic equation. A cubic function can have up to three different roots.

### Example 1:
$$f(x) = x^3-7x^2+14x-8 = (x-1)(x-2)(x-4)$$

coefficients: $a=1$, $b=-7$, $c=14$, $d=-8$

roots: $r_1=1$, $r_2=2$, $r_3=4$

```python
fig, ax = get_base_fig()
ax.plot(x, x**3-7*x**2+14*x-8, color = 'black', label = "$f(x) = x^3-7x^2+14x-8$")
ax.scatter([1,2,4], [0,0,0], color = 'red', s=20, label = "roots: $r_1=1, r_2=2, r_3=4$")
ax.legend(loc = 'upper left') 
ax.set_title("$f(x) = x^3-7x^2+14x-8$")
plt.show()
```

### Example 2:
$$f(x) = -x^3-x^2+2x = -x(x-1)(x+2)$$

coefficients: $a=1$, $b=1$, $c=-2$, $d=0$

roots: $r_1=0, r_2=1, r_3=-2$: 

```python
fig, ax = get_base_fig()
ax.plot(x, -x**3-x**2+2*x, color = 'black', label = "$f(x) = x^3-3x+2$")
ax.scatter([-2, 1, 0], [0,0,0], color = 'red', s=20, label = "roots: $r_1=0, r_2=1, r_3=-2$")
ax.legend(loc = 'upper right') 
ax.set_title("$f(x) = x^3+x^2-2x$")
plt.show()
```

### Example 3:
$$f(x) = x^3+x-x-1 = (x+1)^2(x-1)$$

coefficients: $a=1$, $b=1$, $c=-1$, $d=-1$

roots: $r_1= r_2=-1, r_3=1$

```python
fig, ax = get_base_fig()
ax.plot(x, x**3+x**2-x-1, color = 'black', label = "$f(x) = x^3+x-x-1$")
ax.scatter([-1, -1, 1], [0,0,0], color = 'red', s=20, label = "roots: $r_1=r_2=-1, r_3=1$")
ax.legend(loc = 'upper right') 
ax.set_title("$f(x) = x^3+x-x-1$")
plt.show()
```

### Example 4:
$$f(x) = x^3-3x+2 = (x+2)(x-1)^2$$

coefficients: $a=1$, $b=0$, $c=-3$, $d=2$

roots: $r_1=2, r_2=1, r_3=1$: 

```python
xx = Symbol('xx')
from sympy import solve
print(solve(xx**3 -3*xx + 2,  xx ))
```

```python
fig, ax = get_base_fig()
ax.plot(x, x**3-3*x+2, color = 'black', label = "$f(x) = x^3-3x+2$")
ax.scatter([-2, 1, 1], [0,0,0], color = 'red', s=20, label = "roots: $r_1=-2, r_2=1, r_3=1$")
ax.legend(loc = 'upper right') 
ax.set_title("$f(x) = x^3-3x+2 $")
plt.show()
```

### Example 5:
$$f(x) = x^3-6x^2+12x-8 = (x-2)^3$$

coefficients: $a=1$, $b=-6$, $c=12$, $d=-8$

roots: $r_1 = r_2 = r_3=2$: 

```python
fig, ax = get_base_fig()
ax.plot(x, (x-2)**2, color = 'black', label = "$f(x) = x^3-6x^2+12x-8$")
ax.scatter([2,2,2], [0,0,0], color = 'red', s=20, label = "roots: $r_1=r_2=r_3=2$")
ax.legend(loc = 'upper right') 
ax.set_title("$f(x) =x^3-6x^2+12x-8$")
plt.show()
```

```python
def show_cubic_function_info(a, b, c, d, y_min=-10, y_max=10):
    
    xx = Symbol('xx', real=True)
    roots = solve(a*xx**3 + b*xx**2 + c*xx + d,  xx)
    
    fig, ax = get_base_fig(y_min =y_min, y_max=y_max)
    ax.set_title(f"f(x) = {a} x^3 + {b}x^2 +{c}x + {d}")
    ax.plot(x, a*x**3 + b*x**2 + c*x + d, color = 'black')
    if roots:
        roots_str = ", ".join([f"{root: .2f}" for root in roots])
        ax.scatter(roots, [0]*len(roots), color='red', s=25, label=f'roots: {roots_str} ')       
    ax.legend(loc='lower right')    
    plt.show()
```


# Quartic graphs

A quartic function is the polynomial function defined by a quartic polynomial, i.e., a polynomial of degree four in one or more variables. In this notebook only univariate (single-variable) polynomials are considered. 

A quartic function has the form $$f(x) =f(x) = ax^4 + bx^3 + cx^2 + dx+e$$ where  where $a$, $b$, $c$, $d$ and $e$ are real numbers and $a\neq 0$. If for any value of $x_0$ $f(x_0)=0$, then $x_0$ is called a root of the function $f(x)$. In the other words, if $x_0$ is a root of the function $f(x)$, then the graph of $y = f(x)$ touches or crosses the x-axis at the point $(x_0, 0)$. A quartic function can have at most four different roots. Sometimes, the same root occurs multiple times. In that case, that root is called a multiple root or repeated rood.


A univariate quartic function is usually presented in one of the following forms:


## standard form:
$$f(x) = ax^4 + bx^3 + cx^2 + dx+e$$
where $a$, $b$, $c$ and $d$, $e$ are real numbers and $a\neq 0$.


## factored form
$$ f(x) = a\cdot (x-r_1) \cdot (x-r_2)\cdot (x-r_3)\cdot (x-r_4)$$
$a,r_1, r_2, r_3$ are the roots of the cubic function and the solutions of the corresponding cubic equation. A cubic function can have up to four different roots.

### Example 1:
$$f(x) = x^4 - 6x^3 + 3x^2 + 10x = (x+1)x(x-2)(x-5)$$

coefficients: $a=1$, $b=-6$, $c=3$, $d=10$, $e=0$

roots: $r_1=-1, r_2=0, r_3=2, r_4=5$

```python
fig, ax = get_base_fig(y_min=-50)
ax.plot(x, (x+1)*x*(x-2)*(x-5), color = 'black', label = "$f(x) = x^4 - 6x^3 + 3x^2 + 10x$")
ax.scatter([-1, 0, 2, 5], [0,0,0,0], color = 'red', s=20, label = "roots: $r_1=-1, r_2=0, r_3=2, r_4=5$")
ax.legend(loc = 'upper left') 
ax.set_title("$f(x) =  x^4 - 6x^3 + 3x^2 + 10x$")
plt.show()
```

### Example 2:
$$f(x) = x^4 - 5x^2 + 4 = (x-2)(x-1)(x+1)(x+2)$$

coefficients: $a=1$, $b=0$, $c=-5$, $d=0$, $e=4$

roots: $r_1=-2, r_2=-1, r_3=-1, r_4=2$

```python
fig, ax = get_base_fig()
ax.plot(x, (x-1)*(x+1)*(x-2)*(x+2), color = 'black', label = "$f(x) =x^4 - 5x^2 + 4$")
ax.scatter([-2, -1, 1, 2], [0,0,0,0], color = 'red', s=20, label = "roots: $r_1=-2, r_2=-1, r_3=-1, r_4=2$")
ax.legend(loc = 'lower right') 
ax.set_title("$f(x) = x^4 - 5x^2 + 4$")
plt.show()
```

### Example 2:
$$f(x) = x^4 + 3x^3 -6x^2-28x-24 = (x+2)^3(x-3)$$

coefficients: $a=1$, $b=3$, $c=-6$, $d=-28$, $e=-24$

roots: $r_1=r_2=r_3=-2, r_4 = 3$

```python
fig, ax = get_base_fig(y_min=-70)
ax.plot(x, (x+2)**3*(x-3), color = 'black', label = "$f(x) = x^4 + 3x^3 -6x^2-28x-24$")
ax.scatter([-2, -2, -2, 3], [0, 0, 0, 0], color = 'red', s=20, label = "roots: $r_1=r_2=r_3=-2, r_4 = 3$")
ax.legend(loc = 'lower right') 
ax.set_title("$f(x) = x^4 + 3x^3 -6x^2-28x-24$")
plt.show()
```

### Example 3:
$$f(x) = x^4 + 16x^3 +96x^2 + 256x + 256 = (x+4)^4 $$

coefficients: $a=1$, $b=16$, $c=-96$, $d=256$, $e=256$

roots: $r_1=r_2=r_3=r4 = -4$

```python
fig, ax = get_base_fig()
ax.plot(x, (x+4)**4, color = 'black', label = "$f(x) = x^4 + 16x^3 +96x^2 + 256x + 256 $")
ax.scatter([-4, -4, -4, -4], [0,0, 0,0], color = 'red', s=20, label = "roots: $r_1=r_2=r_3=r4 = -4$")
ax.legend(loc = 'upper right') 
ax.set_title("$f(x) = x^4 + 16x^3 +96x^2 + 256x + 256 $")
plt.show()
```

```python
def show_quartic_function_info(a, b, c, d, e, y_min=-10, y_max=10):
    fig, ax = get_base_fig(y_min =y_min, y_max=y_max)
    ax.set_title(f"f(x) = {a} x^4 + {b}x^3 +{c}x^2 + {d}x+e")
    ax.plot(x, a*x**4 + b*x**3 + c*x**2 + d*x + e, color = 'black')
    ax.legend(loc='lower right')    
    plt.show()
```

# Reciprocal graphs

A reciprocal function has the form: $$f(x) = \frac{a}{x}$$ where $a$ is a constant.

```python
x = x[x!=0]
fig, ax = get_base_fig(y_min=-10, y_max=10, x_min=-5, x_max=5)
ax.plot(x, 1/x, color = 'black', label = r"$f(x) = \frac{1}{x}$")
ax.vlines(x=0, color='white', ymin=-100, ymax = 100, linewidth=5)
ax.vlines(x=0, color='gray', ymin=-100, ymax = 100, linestyles='dashed', alpha=0.7)
ax.legend(loc = 'upper right') 
ax.set_title(r"Reciprocal graphs")
plt.show()
```

```python
x = x[x!=0]

fig, ax = get_base_fig(y_min=-20, y_max=20)
ax.plot(x, 1/x, color = 'black', label = "$f(x) = \frac{1}{x}$")
ax.plot(x, 2/x, color = 'blue', label = "$f(x) = \frac{2}{x}$")
ax.plot(x, 5/x, color = 'green', label = "$f(x) = \frac{5}{x}$")
ax.plot(x, 10/x, color = 'purple', label = "$f(x) = \frac{10}{x}$")
ax.plot(x, 25/x, color = 'red', label = "$f(x) = \frac{25}{x}$")
ax.vlines(x=0, color='white', ymin=-100, ymax = 100, linewidth=5)
ax.vlines(x=0, color='gray', ymin=-100, ymax = 100, linestyles='dashed', alpha=0.7)
ax.legend(loc = 'upper right') 
ax.set_title("Reciprocal graphs")
plt.show()
```
