```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-25, 25, step=0.02)
```

```python
def get_base_fig(x_min = -10, x_max = 10, y_min = -2, y_max = 20):
    fig, ax = plt.subplots(figsize = (11, 6))
    ax.hlines(y=0, color='gray', xmin=-100, xmax = 100, linestyles='dashed', alpha=0.7)
    ax.vlines(x=0, color='gray', ymin=-100, ymax = 100, linestyles='dashed', alpha=0.7)
    ax.set_xlim(x_min, x_max)
    ax.set_ylim(y_min, y_max)
    ax.set_xlabel("x")
    ax.set_ylabel("f(x)")
    ax.grid(alpha=0.3, color = 'gray')                                       
    return fig, ax
```

```python
def get_base_dubleplot(x_min = -10, x_max = 10, y_min = -2, y_max = 20):
    fig, axs = plt.subplots(2, 1, figsize = (11, 8), sharex=True)
    for ax in axs:
        ax.set_xlim(x_min, x_max)
        ax.set_ylim(y_min, y_max)
        ax.set_xlabel("x")
        ax.hlines(y=0, color='gray', xmin=-100, xmax = 100, linestyles='dashed', alpha=0.7)
        ax.vlines(x=0, color='gray', ymin=-100, ymax = 100, linestyles='dashed', alpha=0.7)
        ax.grid(alpha=0.3, color = 'gray')                                       
    axs[0].set_ylabel("f(x)")
    axs[1].set_ylabel("f'(x)")
    return fig, axs
```

```python
def create_tangent_line(f, x0, h, ax, color="red", annotate=False):
    x1 = x0
    x2 = x1+h 
    y1 = f(x1)
    y2 = f(x2)

    a = (y2-y1)/(x2-x1)
    b = (y1*x2 - x1*y2)/(x2-x1)
    
    ax.plot(x, a*x+b, color = color, alpha=0.5, label = f"$f'(x_0)$ = {round(a, 2)}")
    ax.scatter([x1, x2], [y1, y2], color = color, s=10)
    
    if annotate:
        ax.text(x1, y1-1,"$P_1$")
        ax.text(x2, y2-1,"$P_2$")
```

# The gradient

The gradient of a curve is constantly changing. The gradient of a curve at a given point is defined as the gradient of the tangent to the curve at that point. To calculate the gradient of a curve $f(x)$ at the given point $x_0$ we need to find the gradient of a line joining two points: $P_1 = (x_0, f(x_0))$ and $P_2 = (x_0 + \Delta x, f(x_0+\Delta x))$. The smaller value of $\Delta x$, the more accurate is the gradient.


```python
def f(x):
    return x**3-3*x+2

def f_prim(x):
    return 3*x**2 - 3
```

```python
y = f(x)
x0 = 2.1

h = 0.5

fig, ax = get_base_fig(x_min = -3, x_max =3, y_min = -5, y_max = 15)
fig.suptitle(f"Tangent line to $f(x)=x^3-3x+2$ \nat the point $x_0 = {x0}$")
create_tangent_line(f, x0, h, ax, annotate=True)
ax.plot(x, y, color = 'black', label = "$f(x) = x^3-3x+2$")
ax.legend(loc = 'upper left')
plt.show()



h = 0.05

fig, ax = get_base_fig(x_min = -3, x_max =3, y_min = -5, y_max = 15)
fig.suptitle(f"Tangent line to $f(x)=x^3-3x+2$ \nat the point $x_0 = {x0}$")
create_tangent_line(f, x0, h, ax, annotate=True)
ax.plot(x, y, color = 'black', label = "$f(x) = x^3-3x+2$")
ax.legend(loc = 'upper left')
plt.show()



h = 0.005

fig, ax = get_base_fig(x_min = -3, x_max =3, y_min = -5, y_max = 15)
fig.suptitle(f"Tangent line to $f(x)=x^3-3x+2$ \nat the point $x_0 = {x0}$")
create_tangent_line(f, x0, h, ax, annotate=True)
ax.plot(x, y, color = 'black', label = "$f(x) = x^3-3x+2$")
ax.legend(loc = 'upper left')
plt.show()
```

# The gradinet c.d

The gradient of a function $f(x)$ tells how fast $f(x)$ is changing. Moreover, it tells, whether $f(x)$ is increasing of decreasing at the given interval.

* If the gradient of a function $f(x)$ is negative on some interval $[a, b]$, i.e. $f'(x) < 0$ for x such that $a < x < b$, then the function f(x) is increasing on the interval $[a, b]$.

* If the gradient of a function $f(x)$ is positive on some interval $[a, b]$, i.e. $f'(x) > 0$ for x such that $a < x < b$, then the function f(x) is increasing on the interval $[a, b]$.

* If the gradient of a function $f(x)$ is equal to zero at some point $x_0$, i.e. $f'(x_0) = 0$, then the function f(x) has a local extremum (minimum, maximum or point of inflection) at $x_0$

```python
y = f(x)
h = 0.001
x0 = 1.9

fig, ax = get_base_fig(x_min = -5, x_max =5, y_min = -5, y_max = 15)
fig.suptitle(f"Tangent line to $f(x)=x^3-3x+2$ \nat the point $x_0 = {x0}$")
create_tangent_line(f, x0, h, ax)
ax.plot(x, y, color = 'black', label = "$f(x) = x^3-3x+2$")
ax.legend(loc = 'upper left')
plt.show()
```

# Mathematical formula for  the gradient


The gradient function, or derivative, of the curve $y = f(x)$ at a point $x_0$ is defined as the limit$^1$:
$$ f'(x) = \frac{dy}{dx} = \lim _{h\to 0}{\frac {f(x_0+h)-f(x_0)}{h}}$$ 


$^1$ It is necessary to check, if the given function $f(x)$ is differentiable at the point $x_0$, i.e. the domain of $f(x)$ contains an open interval I containing $x_0$, and the limit exists.



## Caluclating derivatives from First Principles:


### Constant function:

$ f(x) = a_0 $


$$\begin{equation*}
  \left\{
    \begin{aligned}
       f'(x) & = \lim _{h\to 0}{\frac {f(x_0+h)-f(x_0)}{h}} \\
       f'(x) & = \lim _{h\to 0}{\frac {a_0 - a_0}{h}} \\
       f'(x) & = \lim _{h\to 0}{\frac {0}{h}}  = 0 
    \end{aligned}
  \right.
\end{equation*}$$


### linear function:

$ f(x) = a_1 \cdot x + a_0 $

$ f(x_0) = a_1 \cdot x_0 + a_0 $

$ f(x_0+h) = a_1 \cdot (x_0+h) + a_0$

$$\begin{equation*}
  \left\{
    \begin{aligned}
       f'(x) & = \lim _{h\to 0}{\frac {f(x_0+h)-f(x_0)}{h}} \\
       f'(x) & = \lim _{h\to 0}{\frac {a_1 \cdot (x_0+h) + a_0 - ( a_1 \cdot x_0 + a_0)}{h}} \\
       f'(x) & = \lim _{h\to 0}{\frac {a_1\cdot x_0 + a_1\cdot h + a_0 - a_1\cdot x_0 - a_0)}{h}} \\
       f'(x) & = \lim _{h\to 0}{\frac {a_1 \cdot h}{h}} = a_1 
    \end{aligned}
  \right.
\end{equation*}$$



### quadratic function:

$ f(x) = a_2 \cdot x^2 $

$ f(x_0) = a_2 \cdot x_0^2  $

$ f(x_0+h) = a_2 \cdot (x_0+h)^2$

$$\begin{equation*}
  \left\{
    \begin{aligned}
       f'(x) & = \lim _{h\to 0}{\frac {f(x_0+h)-f(x_0)}{h}} \\
       f'(x) & = \lim _{h\to 0}{\frac {a_2 \cdot (x_0+h)^2 - a_2 \cdot x_0^2}{h}} \\
       f'(x) & = \lim _{h\to 0}{\frac {a_2 \cdot x_0^2 + 2 \cdot a_2 \cdot x_0\cdot h + a_2 h^2 - a_2 \cdot x_0^2}{h}} \\
       f'(x) & = \lim _{h\to 0}{\frac {2\cdot a_2 \cdot x_0\cdot h + a_2 h^2}{h}}\\
       f'(x) & = \lim _{h\to 0}  2\cdot a_2 \cdot x_0 + a_2 h = a_2 \cdot x + \mathcal{O}(h)\\
    \end{aligned}
  \right.
\end{equation*}$$



### cubic function (TO DO):

$ f(x) = a_3 \cdot x^3 $

$ f(x_0) = a_3 \cdot x_0^3  $

$ f(x_0+h) = a_3 \cdot (x_0+h)^3$


```python
def get_formatted_function(function):
    function = function.replace("**", "^").replace("*", "\cdot ")
    return "$" + function + "$"

def show_function_derivative(input_function, input_derivative, x0, title, show_plot=True):
 
    f = lambda x: eval(input_function, {}, {"x": x})
    f_prim =  lambda x: eval(input_derivative, {}, {"x": x}) 
    
    fig, axs = get_base_dubleplot(x_min = -10, x_max =10, y_min = -10, y_max = 15)
    fig.suptitle(title)
    create_tangent_line(f, x0, h=0.0001, ax=axs[0])
    axs[1].scatter([x0], [f_prim(x0)], color="red", s=15)
    axs[0].plot(x, f(x), color = 'blue', label = "f(x) =" + get_formatted_function(input_function)  )
    axs[1].plot(x, f_prim(x), color = 'blue', label = "f'(x) =" + get_formatted_function(input_derivative) )
    axs[0].legend(loc = 'upper left')
    axs[1].legend(loc = 'upper left')
    
    if show_plot:
        plt.show()
    else:
        return fig, axs
```

```python
input_function = "0.5 * x + 1"
input_derivative = "0.5 * x**0" 
x0 = 2.1

title = "The derivative of a linear function"
show_function_derivative(input_function, input_derivative, x0, title);
```

```python
input_function = "0.25 * x**2 - x + 3"
input_derivative = "0.25 * 2 * x - 1" 
x0 = 2

title = "The derivative of a quadratic function"
show_function_derivative(input_function, input_derivative, x0, title);
```

# Stationary points

If the gradient of a function $f(x)$ is equal to zero at some point $x_0$, i.e. $f'(x_0) = 0$, then the function f(x) has at $x_0$ a local extremum:
* local minimum
* local maximum
* point of inflection


```python
input_function = "x**4 - 3*x**2"
input_derivative = "4*x**3 - 6*x" 
x0 = 0

title = "Local maximum"
fig, axs = show_function_derivative(input_function, input_derivative, x0, title, show_plot=False)
```

```python
input_function = "x**4 - 4*x**2"
input_derivative = "4*x**3 - 8*x" 
x0 = np.sqrt(2)

title = "Local minimum"
fig, axs = show_function_derivative(input_function, input_derivative, x0, title, show_plot=False)
```

```python
input_function = "x**3 - 3"
input_derivative = "3*x**2" 
x0 = 0

title = "Inflection point"
fig, axs = show_function_derivative(input_function, input_derivative, x0, title, show_plot=False)
```
