```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 1001)
```

```python
def get_base_fig(x_min = -8, x_max = 8, y_min = -10, y_max = 25, step=2):
    fig, ax = plt.subplots(figsize = (11, 6))
    ax.hlines(y=0, color='gray', xmin=-100, xmax = 100, linestyles='dashed', alpha=0.7)
    ax.vlines(x=0, color='gray', ymin=-100, ymax = 100, linestyles='dashed', alpha=0.7)
    ax.set_xlim(x_min, x_max)
    ax.set_ylim(y_min, y_max)
    ax.set_xlabel('$x$')
    ax.set_ylabel(r"f(x)")
    ax.set_xticks(np.arange(x_min, x_max, step = 1))
    ax.set_yticks(np.arange(y_min, y_max, step = step))
    ax.grid(alpha=0.3, color = 'gray')                                       
    return fig, ax
```

# System of nonlinear equations

In mathematics, a system of nonlinear equations is a collection of one or more linear equations involving the same variables. A solution to a nonlinear system is an assignment of values to the variables such that all the equations are simultaneously satisfied. In this notebook a graphical method for solving system of linear equations for two 
variables is shown. 

### Example 1:

Find the coordinates of the point of intersection:
$$\begin{equation*}
  \left\{
    \begin{aligned}
      & y = x^2 -4x + 7\\
      & y = 3x + 1
    \end{aligned}
  \right.
\end{equation*}$$

$$y = x^2-4x + 7 = 3x + 1$$ 

$$0 = x^2-7x+6 = (x-1)(x-6)$$ 

The above quadratic function has two real solutions:

$$\begin{equation*}
  \left\{
    \begin{aligned}
      & x_1 = 1 \rightarrow y_1 = 4 \\
      & x_2 = 6 \rightarrow y_2 = 19
    \end{aligned}
  \right.
\end{equation*}$$


```python
fig, ax = get_base_fig(y_min=-15)
ax.plot(x, 3*x+1, color = 'blue', label = r"$f(x) = 3x + 1$")
ax.plot(x, x**2 - 4*x + 7, color = 'green', label = r"$f(x) = x^2 -5x + 7$")
ax.scatter([1, 6], [4, 19], color = 'red', s=20,  
           label = r"$p_{intersection}= (1, 4)$" + "\n" + r"$p_{intersection}= (6,19)$" )
ax.legend(loc = 'lower right') 
plt.show()
```

### Example 2:

Find the coordinates of the point of intersection:
$$\begin{equation*}
  \left\{
    \begin{aligned}
      & y = x^2\\
      & y = 2x^2-2x-12
    \end{aligned}
  \right.
\end{equation*}$$

$$y = x^2 = 2x^2 -2x-12$$ 

$$0 = x^2-2x-12$$ 

The above quadratic function has two real solutions:

$$\begin{equation*}
  \left\{
    \begin{aligned}
      & x_1 = 1- \sqrt(13) \rightarrow y_1 = \left( 1- \sqrt(13)\right)^2\\
      & x_2 = 1+ \sqrt(13) \rightarrow y_2 = \left( 1+ \sqrt(13)\right)^2
    \end{aligned}
  \right.
\end{equation*}$$


```python
fig, ax = get_base_fig(y_min=-15)
ax.plot(x, x**2, color = 'blue', label = r"$f(x) = x^2$")
ax.plot(x, 2*(x-3)*(x+2), color = 'green', label = r"$f(x) = 2x^2-2x-12$")
ax.scatter([1-np.sqrt(13), 1+np.sqrt(13)], [(1-np.sqrt(13))**2, (1+np.sqrt(13))**2], color = 'red', s=20,  
           label = r"$p_{intersection}= (1-\sqrt{13}, (1-\sqrt{13})^2)$" + "\n" + r"$p_{intersection} = (1+\sqrt{13}, (1+\sqrt{13})^2)$")
ax.legend(loc = 'lower right') 
plt.show()
```

### Example 3:

Find the coordinates of the point of intersection:
$$\begin{equation*}
  \left\{
    \begin{aligned}
      & 2x + 2y = 7\\
      & x^2-4y^2=8
    \end{aligned}
  \right.
\end{equation*}$$

Lines:

$$y = 7/2 - x$$ 
$$ y = \pm \frac{\sqrt{8-x^2}}{2}$$


Solution:

From the first exquation we have: $x = \frac{7}{2} -y$
We insert that to the second equation:
$$ \left(\frac{7}{2}-y\right)^2 - 4y^2 = 8 $$

$$ y^2 - \frac{14}{2}y + \frac{49}{4} - 4y^2 = 8 $$
$$ -3y^2 - 7y + \frac{17}{4}=0 $$
$$ 12y^2 +28y -17=0 $$

We solve the quadratic equation: $\Delta = 784 -4\cdot (-17)\cdot 12 = 1600$


$$\begin{equation*}
  \left\{
    \begin{aligned}
      & y_1 = \frac{-28+40}{2\cdot12} = \frac{1}{2} ~~\rightarrow ~~x_1 = 3\\
      & y_2= \frac{-28-40}{2\cdot12} = -\frac{17}{6} ~~\rightarrow ~~ x_2 = \frac{19}{3}
    \end{aligned}
  \right.
\end{equation*}$$

```python
fig, ax = get_base_fig(y_min=-15)
ax.plot(x, 7/2-x, color = 'blue', label = r"$f(x) = 2x + 2y = 7$")
ax.plot(x, np.sqrt(x**2-8)/2, color = 'green', label = r"$f(x) = x^2 -4y^2 = 8$")
ax.plot(x, -np.sqrt(x**2-8)/2, color = 'green')

ax.scatter([3, 19/3], [0.5, -17/6], color = 'red', s=20,  
           label = r"$p_{intersection}= (1, 4)$" + "\n" + r"$p_{intersection}= (6,19)$" )
ax.legend(loc = 'lower right') 
ax.set_ylim(-10, 10)
plt.show()
```

### Example 4:

Find the coordinates of the point of intersection:
$$\begin{equation*}
  \left\{
    \begin{aligned}
      & y = \frac{4}{x^2}\\
      & y = x^2\cdot(x-3)
    \end{aligned}
  \right.
\end{equation*}$$

$$y = \frac{4}{x^2} = x^2\cdot(x-3)$$ 

$$4 = x^4\cdot(x-3)$$ 
$$0 = x^4\cdot(x-3) -4$$ 

```python
fig, ax = get_base_fig()
ax.plot(x, x**2*(x-3), color = 'blue', label = r"$f(x) = x^2\cdot (x-3)$")
ax.plot(x, 4/x**2, color = 'green', label = r"$f(x) = \frac{4}{x^2}$")
ax.scatter([3.0464], [0.4310], color = 'red', s=20, label = r"$p_{intersection}: (3.0464, 0.4310)$")
ax.legend(loc = 'upper left') 
plt.show()
```

### Example 5:

Find the coordinates of the point of intersection:
$$\begin{equation*}
  \left\{
    \begin{aligned}
      & x+y=11 \\
      & x\cdot y =30 
    \end{aligned}
  \right.
\end{equation*}$$


Solution:
From the first equation: 
$$x = 11-y$$

Inserting $x$ to the second equation gives:
$$x\cdot y = (11-y)\cdot y = 30$$ 
$$11y - y^2 = 30$$ 
$$0 = y^2 - 11y + 30$$

Using factorization we can find the solutions of the above quadratic function
$$0 = (y -5)(y-6)$$
$$ y_1 = 5, y_2 = 6$$

for $y_1 = 5$ we have:
$$ x_1= 11-y_1 = 11-5 = 6$$

for $y_2 = 6$ we have:
$$ x_2 = 11-y_2 = 11-6 = 5$$

There are two sets of solutions:
$$\begin{equation*}
  \left\{
    \begin{aligned}
      & x_1 = 5 \\
      & y_1 = 6
    \end{aligned}
  \right.
\end{equation*}$$

$$\begin{equation*}
  \left\{
    \begin{aligned}
      & x_2 = 5 \\
      & y_2 = 6
    \end{aligned}
  \right.
\end{equation*}$$

```python
fig, ax = get_base_fig(x_min=1, x_max=9, y_min=1, y_max=9)
x = np.linspace(1, 10, 1001)
ax.plot(x, 11-x, color = 'blue', label = r"$x+y=11$")
ax.plot(x, 30/x, color = 'green', label = r"$x \cdot y =30$")
ax.scatter([5,6], [6,5], color='red', s=25, label = r"$p_{interection}=(5,6)$" + "\n" + r"$p_{interection}=(6,5)$")
ax.legend(loc = 'center right')
plt.show()
```

### Example 6:

Find the coordinates of the point of intersection:
$$\begin{equation*}
  \left\{
    \begin{aligned}
      & 2x+y=1 \\
      & x^2 + y^2 =1 
    \end{aligned}
  \right.
\end{equation*}$$


Solution:
From the first equation: 
$$y = 1-2x$$

Inserting $x$ to the second equation gives:
$$x^2 + y^2 = x^2 + (1-2x)^2 = 1$$ 
$$x^2 + 4x^2 - 4x +1 =1$$ 
$$5x^2 - 4x +1 = 1$$
$$5x^2 - 4x = 0$$

Using factorization we can find the solutions of the above quadratic function
$$x(5x-4) = 0$$
$$x_1=0, x_2 = \frac{4}{5}$$

for $x_1 = 0$ we have:
$$ y_1 = 1-2x_1 = 1$$

for $x_2 = \frac{4}{5}$ we have:
$$ y_2 = 1-2x_2 = 1-\frac{8}{5} = 1-1.5 = -0.6$$

There are two sets of solutions:
$$\begin{equation*}
  \left\{
    \begin{aligned}
      & x_1 = 0 \\
      & y_1 = 1
    \end{aligned}
  \right.
\end{equation*}$$

$$\begin{equation*}
  \left\{
    \begin{aligned}
      & x_2 = 0.8 \\
      & y_2 = -0.6
    \end{aligned}
  \right.
\end{equation*}$$

```python
fig, ax = get_base_fig(x_min=-1.5, x_max=1.5, y_min=-1.5, y_max=1.5, step=0.25)
ax.plot(x, 1-2*x, color = 'blue', label = r"$2x+y=1$")
ax.plot(x, np.sqrt(1-x**2), color = 'green', label = r"$x^2+y^2=1$")
ax.plot(x, -np.sqrt(1-x**2), color = 'forestgreen')

ax.scatter([0,0.8], [1, -0.6], color='red', s=25, label = r"$p_{interection}=(0,1)$" + "\n" + r"$p_{interection}=(0.8, -0.6)$")
ax.legend(loc = 'upper right')
plt.show()
```

# Parametric equations

### Example 1:

On the same diagram sketch the curves with equations $y =  k \cdot x^2$ and $y = x+m$, where $k$ and $m$
are real constants. State the number of real solutions to the set of equations:

$$\begin{equation*}
  \left\{
    \begin{aligned}
      & y = k \cdot x^2 \\
      & y = x+m
    \end{aligned}
  \right.
\end{equation*}$$


$$ y = k \cdot x^2 = x+m$$
$$ 0 = k \cdot x^2 -x-m $$
$$ \Delta = b^2-4ac = (-1)^2 -4\cdot k\cdot (-m) = 1+4\cdot k\cdot m $$

* if $\Delta > 0$ or $k\cdot m > -\frac{1}{4}$ then there are two distinct intersection points
* if $\Delta = 0$ or $k\cdot m = -\frac{1}{4}$  then there is one intersection point
* if $\Delta < 0$ or $k\cdot m < -\frac{1}{4}$ then there are no intersection points

```python
k_list = [0.25, 0.5, 1, 2]
m_list = [-3, -1, 1, 3]

fig, ax = get_base_fig(y_min=-10, y_max=10)

for i, k in enumerate(k_list):
    ax.plot(x, k*x**2, color = 'green', alpha=1-2*i/10, label = f"a={k}")

for i, m in enumerate(m_list):
    ax.plot(x, x+m, color = 'blue', alpha=1-2*i/10, label = f"m={m}")

ax.legend(loc = 'upper left') 
plt.show()
```

### Example 2:

On the same diagram sketch the curves with equations $y = \frac{a}{x}$ and $y = x^2(3x − b)$, where $a$ and $b$
are positive constants. 
State the number of real solutions to the set of equations:

$$\begin{equation*}
  \left\{
    \begin{aligned}
      & y = \frac{a}{x}\\
      & y = x^2\cdot(x-b)
    \end{aligned}
  \right.
\end{equation*}$$

```python
a_list = [1, 5, 10, 15, 20]
b_list = [1, 2, 3, 4, 5]

fig, ax = get_base_fig(y_min=-20, y_max=20)

for i, a in enumerate(a_list):
    ax.plot(x, a/x, color = 'green', alpha=1-i/10, label = f"a={a}")

    ax.vlines(x=0, color='white', ymin=-100, ymax = 100, linewidth=5)
ax.vlines(x=0, color='gray', ymin=-100, ymax = 100, linestyles='dashed', alpha=0.7)

for i, b in enumerate(b_list):
    ax.plot(x, x**2*(x-b), color = 'blue', alpha=1-i/10, label = f"b={b}")

ax.legend(loc = 'upper left') 
plt.show()
```

### Example 4:


On the same diagram sketch the curves with equations $y = x+b$ and $r^2 = x^2 + y^2)$, where $a$ and $b$
are positive constants. 
State the number of real solutions to the set of equations:

$$\begin{equation*}
  \left\{
    \begin{aligned}
      & y = x + b  \\
      & x^2 + y^2 = r^2
    \end{aligned}
  \right.
\end{equation*}$$



```python
r_list = [0.25, 0.5, 1, 2]
b_list = [1, 2, 3, 4, 5]

fig, ax = get_base_fig(y_min=-10, y_max=10)

for i, r in enumerate(r_list):
    ax.plot(x, np.sqrt(r**2-x**2), color = 'green', label = r"$x^2+y^2=r$")
    ax.plot(x, -np.sqrt(r**2-x**2), color = 'forestgreen')


for i, b in enumerate(b_list):
    ax.plot(x, x+b, color = 'blue', alpha=1-i/10, label = f"b={b}")

ax.legend(loc = 'upper left') 
plt.show()
```
