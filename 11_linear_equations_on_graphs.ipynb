```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import numpy as np

x = np.linspace(-5, 10, 201)
```

```python
def get_base_fig(x_min = -5, x_max = 10, y_min = -2, y_max = 8, step=1):
    fig, ax = plt.subplots(figsize = (11, 6))
    ax.hlines(y=0, color='gray', xmin=-10, xmax = 100, linestyles='dashed', alpha=0.7)
    ax.vlines(x=0, color='gray', ymin=-10, ymax = 100, linestyles='dashed', alpha=0.7)
    ax.set_xlim(x_min, x_max)
    ax.set_ylim(y_min, y_max)
    ax.set_xlabel('$x$')
    ax.set_ylabel(r"f(x)")
    ax.set_xticks(np.arange(x_min, x_max, step = 1))
    ax.set_yticks(np.arange(y_min, y_max, step = step))
    ax.grid(alpha=0.3, color = 'gray')                                       
    return fig, ax
```

# System of linear equations

In mathematics, a system of linear equations (or linear system) is a collection of one or more linear equations involving the same variables. A solution to a linear system is an assignment of values to the variables such that all the equations are simultaneously satisfied. In this notebook a graphical method for solving system of linear equations for two and three variables is shown. 

For a system of two linear equations graphical method requires drawing two lines corresponding to the equations and finding the point of intersection for those two lines. It may happen that there is no point of intersection, and the system of two linear equations has no solutions or there are infinitelly many intersection points and the system of two linear equations has infinitelly many solutions


For a system of three linear equations graphical method requires drawing three planes corresponding to the equations and finding point of intersection of those planes

# Linear simultaneous equations for two variables

### Example 1:

Find the coordinates of the point of intersection:
$$\begin{equation*}
  \left\{
    \begin{aligned}
      & y = 3x-5 \\
      & y = 3-x
    \end{aligned}
  \right.
\end{equation*}$$


Solution (using substitution method):
$$y = 3x-5 = 3-x$$ 
$$4x = 8$$ 
$$x = 2 $$

Now, from the first equation:
$$y = 3-x = 3-2 =1$$

so:

$$\begin{equation*}
  \left\{
    \begin{aligned}
      & x =2 \\
      & y =1
    \end{aligned}
  \right.
\end{equation*}$$

```python
fig, ax = get_base_fig()
ax.plot(x, 3*x-5, color = 'blue', label = r"$y = 3x-5$")
ax.plot(x, 3-x, color = 'green', label = r"$y = 3-x$")
ax.scatter([2], [1], color='red', s=25, label = r"$p_{intersection}=(2,1)$")
ax.legend(loc = 'center right')
plt.show()
```

### Example 2:

Find the coordinates of the point of intersection:
$$\begin{equation*}
  \left\{
    \begin{aligned}
      & y = 2x-7 \\
      & y = 8-3x
    \end{aligned}
  \right.
\end{equation*}$$


Solution (using substitution method):
$$y = 2x-7 = 8-3x$$ 
$$5x = 15$$ 
$$x=3 $$

Now, from the first equation:
$$y = 2x -7 = 2\cdot 3-7= -1$$

so:

$$\begin{equation*}
  \left\{
    \begin{aligned}
      & x = 3 \\
      & y = -1
    \end{aligned}
  \right.
\end{equation*}$$

```python
fig, ax = get_base_fig()
ax.plot(x, 2*x-7, color = 'blue', label = r"$y = 2x-7$")
ax.plot(x, 8-3*x, color = 'green', label = r"$y = 8-3x$")
ax.scatter([3], [-1], color='red', s=25, label = r"$p_{intersection}=(3,-1)$")
ax.legend(loc = 'center right')
plt.show()
```

### Example 3:

Find the coordinates of the point of intersection:
$$\begin{equation*}
  \left\{
    \begin{aligned}
      & y = 3x+2 \\
      & 3x+y+1=0 
    \end{aligned}
  \right.
\end{equation*}$$


Solution (using substitution method):
$$y = 3x+2 = -1-3x$$ 
$$6x = -3$$ 
$$x = -0.5$$

Now, from the first equation:
$$y = 3x+2 = 3\cdot 0.5 + 2 = 0.5$$

so:

$$\begin{equation*}
  \left\{
    \begin{aligned}
      & x = -0.5 \\
      & y = 0.5
    \end{aligned}
  \right.
\end{equation*}$$

```python
fig, ax = get_base_fig()
ax.plot(x, 3*x+2, color = 'blue', label = r"$y = 3x+2$")
ax.plot(x, -3*x-1, color = 'green', label = r"$3x+y+1=0$")
ax.scatter([-0.5], [0.5], color='red', s=25, label = r"$p_{intersection}=(-0.5, 0.5)$")
ax.legend(loc = 'center right')
plt.show()
```

### Example 4:

Find the coordinates of the point of intersection:
$$\begin{equation*}
  \left\{
    \begin{aligned}
      & y = 2x-7 \\
      & y = 2x+1
    \end{aligned}
  \right.
\end{equation*}$$

This system of equations has no solutions. Lines given by $= 2x-7 $ and $= 2x+1$ have no point of intersection as they are parallel

```python
fig, ax = get_base_fig()
ax.plot(x, 2*x-7, color = 'blue', label = r"$y = 2x-7$")
ax.plot(x, 2*x+1, color = 'green', label = r"$y=2x+1$")
ax.legend(loc = 'center right')
plt.show()
```

# Linear simultaneous equations in for three variables

### Example 1:

Find the coordinates of the point of intersection:

$$\begin{equation*}
  \left\{
    \begin{aligned}
      & 2x-y+z = -4 \\
      & 8x+2y-5z = -10  \\
      & 4x+y+z = 2
    \end{aligned}
  \right.
\end{equation*}$$


$$\begin{equation*}
  \left\{
    \begin{aligned}
      & z = -4 + y -2x \\
      & z = 1.6x + 0.4y +2 \\
      & z = 2 - 4x -y
    \end{aligned}
  \right.
\end{equation*}$$

Solution:

$$\begin{equation*}
  \left\{
    \begin{aligned}
      & x = -1 \\
      & y = 4 \\
      & z = 2 \\
    \end{aligned}
  \right.
\end{equation*}$$

```python
x = np.linspace(-4, 6, 201)
y = np.linspace(-4, 6, 201)
x, y = np.meshgrid(x, y)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.plot_wireframe(x, y, -2*x + y -4, color = 'blue', alpha=0.2, label=r'$z = -4 + y -2x$')
ax.plot_wireframe(x, y, (8*x+2*y+10)/5, color = 'green', alpha=0.2, label=r'$z = 1.6x + 0.4y +2$')
ax.plot_wireframe(x, y, -4*x-y+2, color = 'purple', alpha=0.2, label=r'z = 2 - 4x -y')
ax.scatter([-1], [4], [2], color='red', s=30)

ax.set_xlabel('x');
ax.set_ylabel('y');
ax.set_zlabel('z');

ax.legend(loc = 'center right', bbox_to_anchor=(1, 1))
ax.view_init(30, 45, 0)
```

### Example 2:

Find the coordinates of the point of intersection:

$$\begin{equation*}
  \left\{
    \begin{aligned}
      & x-2y+3z = -7 \\
      & 3x+y+4z=5  \\
      & 2x + 5y + z = 18
  \end{aligned}
  \right.
\end{equation*}$$


$$\begin{equation*}
  \left\{
    \begin{aligned}
      & z = \frac{-7-x+2y}{3} \\
      & z = \frac{5-3x-y}{4} \\
      & z =18 - 2x - 5y
    \end{aligned}
  \right.
\end{equation*}$$

Solution:

$$\begin{equation*}
  \left\{
    \begin{aligned}
      & x = 2 \\
      & y = 3 \\
      & z = -1 \\
    \end{aligned}
  \right.
\end{equation*}$$

```python
x = np.linspace(-4, 6, 201)
y = np.linspace(-4, 6, 201)
x, y = np.meshgrid(x, y)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.plot_wireframe(x, y, (-7-x+2*y)/3, color = 'blue', alpha=0.2, label=r'$z = \frac{-7-x+2y}{3}$')
ax.plot_wireframe(x, y, (5-y-3*x)/4, color = 'green', alpha=0.2, label=r'$z = \frac{5-3x-y}{4}$')
ax.plot_wireframe(x, y, 18-2*x-5*y, color = 'purple', alpha=0.2, label=r'z =18 - 2x - 5y')
ax.scatter([2], [3], [-1], color='red', s=30)

ax.set_xlabel('x');
ax.set_ylabel('y');
ax.set_zlabel('z');

ax.legend(loc = 'center right', bbox_to_anchor=(1,1))
ax.view_init(15, 15, 0)
```
