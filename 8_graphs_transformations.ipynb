```python
import matplotlib.pyplot as plt
import numpy as np

epsilon = 0.0001
x = np.arange(-25, 25, step=0.02) + epsilon
```

```python
def get_base_fig(x_min = -8, x_max = 8, y_min = -10, y_max = 25):
    fig, ax = plt.subplots(figsize = (11, 6))
    ax.hlines(y=0, color='gray', xmin=-100, xmax = 100, linestyles='dashed', alpha=0.7)
    ax.vlines(x=0, color='gray', ymin=-100, ymax = 100, linestyles='dashed', alpha=0.7)
    ax.set_xlim(x_min, x_max)
    ax.set_ylim(y_min, y_max)
    ax.set_xlabel("x")
    ax.set_ylabel("f(x)")
    ax.grid(alpha=0.3, color = 'gray')                                       
    return fig, ax
```

# Graph transformations


Graph transformation is the process by which an existing graph, or graphed equation, is modified to produce a variation of the proceeding graph.


* translations
* reflections
* scaling: strething or compressing
* rotations

## Graph translations

#### Vertical translations:
The graph $y=f(x) + a$ is a translation of the graph $y=f(x)$ by the vector $ {{0}\choose{a}}$

#### Horizontal translations:
The graph $y=f(x+a)$ is a translation of the graph $y=f(x)$ by the vector $ {{-a}\choose{0}}$

The graph $y=f(x-a)$ is a translation of the graph $y=f(x)$ by the vector $ {{a}\choose{0}}$

```python
def translate_graph(x, function, translation_vector, **kwargs):
    y_original = function(x)
    y_translated = function(x-translation_vector[0]) + translation_vector[1]
    x1 = translation_vector[0]
    y1 = translation_vector[1] + function(0+epsilon)
    y2 = function(0+epsilon)
    
    fig, ax = get_base_fig(**kwargs)
    ax.plot(x, y_original, color='black', label = "original")
    ax.plot(x, y_translated, color='green', label = "translated")
    ax.annotate("", xy=[x1,y1], xytext=(0, y2), arrowprops=dict(width=1, ec='red', fc='red'))
    ax.legend(loc="lower right")
    plt.show()
```

```python
function = lambda x: x**2
translate_graph(x, function, translation_vector=[0,1], x_min = -4, x_max = 4, y_min = -1, y_max = 12)
```

```python
function = lambda x: x**2
translate_graph(x, function, translation_vector=[3,0], x_min = -4, x_max = 4, y_min = -1, y_max = 12)
```

```python
function = lambda x: x**2  ## (x-2)**2
translate_graph(x, function, translation_vector=[-2,0], x_min = -4, x_max = 4, y_min = -1, y_max = 12)
```

```python
function = lambda x: (x-1)*(x+2)
translation_vector = [-1,2]
translate_graph(x, function, translation_vector, x_min = -4, x_max = 4, y_min = -3, y_max = 12)
```

```python
function = lambda x: np.sin(x)
translation_vector = [-1,2]
translate_graph(x, function, translation_vector, x_min = -4, x_max = 4, y_min = -1.5, y_max = 3.5)
```

## Graph reflections

#### Vertical reflections:
The graph of $y = –f(x)$ is a reflection of the graph of $y = f(x)$ in the x-axis

#### Horizontal reflections:
The graph of $y = f(–x)$ is a reflection of the graph of $y = f(x)$ in the y-axis.

```python
def reflect_graph(x, function, axis, **kwargs):
    y_original = function(x)
    
    if axis == 0:
        y_reflected = (-1)*function(x)
    elif axis == 1:
        y_reflected = function(-x)

    fig, ax = get_base_fig(**kwargs)
    ax.plot(x, y_original, color='black', label = "original")
    ax.plot(x, y_reflected, color='green', label = "reflected")
    ax.legend(loc="lower right")
    plt.show()
```

```python
function = lambda x: (x-1)*(x-2)
reflect_graph(x, function, axis=0, x_min = -10, x_max = 10, y_min = -10, y_max = 10)
```

```python
function = lambda x: (x-1)*(x-2)
reflect_graph(x, function, axis=1, x_min = -10, x_max = 10, y_min = -10, y_max = 10)
```

```python
function = lambda x: np.sin(x)
reflect_graph(x, function, axis=1, x_min = -10, x_max = 10, y_min = -10, y_max = 10)
```

```python
function = lambda x: np.cos(x)
reflect_graph(x, function, axis=1, x_min = -10, x_max = 10, y_min = -10, y_max = 10)
```

```python
function = lambda x: np.cos(x+ np.pi/4)
reflect_graph(x, function, axis=1, x_min = -10, x_max = 10, y_min = -10, y_max = 10)
```

### Stretching and compressing graphs

#### Stretching/compressing graphs in the vertical direction
The graph of $y = a\cdot f(x)$ is a stretch of the graph $y = f(x)$ by a scale factor of $a$ in the vertical
direction.

#### Stretching/compressing graphs in the horizontal direction
The graph of $y = f(a\cdot x)$ is a stretch of the graph $y = f(x)$ by a scale factor of $\frac{1}{a}$ in the
horizontal direction.

```python
def scale_graph(x, function, scaling_factor, direction,**kwargs):
    
    y_original = function(x)
    if direction == 0:
        y_stretched = scaling_factor * function(x)
    elif direction == 1:
        y_stretched = function(x*scaling_factor)
    
    fig, ax = get_base_fig(**kwargs)
    ax.plot(x, y_original, color='black', label = "original")
    ax.plot(x, y_stretched, color='green', label = "reflected")
    ax.legend(loc="lower right")
    plt.show()
```

```python
function = lambda x: np.sin(x)
scaling_factor = 2
scale_graph(x, function, scaling_factor, direction=0, x_min = -4, x_max = 4, y_min = -1.5, y_max = 3.5)
```

```python
function = lambda x: np.sin(x)
scaling_factor = 2
scale_graph(x, function, scaling_factor, direction=1, x_min = -4, x_max = 4, y_min = -1.5, y_max = 3.5)
```

```python
input_function = input("Enter a function f(x) = ")
function = lambda x: eval(input_function, {}, {"x": x})

scaling_factor = int(input("Enter the scaling factor: "))

direction = int(input("Enter the direction: 0 for horizontal stretching, 1 for vertical stretching: "))
scale_graph(x, function, scaling_factor, direction=direction, x_min = -10, x_max = 10, y_min = -10, y_max = 10)
```
